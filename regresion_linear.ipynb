```python
#la linea
datos['metro'].min()
```

```python
# parametros de la recta
m = 10
b = 300
```

```python
# puntos de la recta
x = np.linspace(0,datos['metro'].max(),100)
y = m*x+b
```

```python
# grafica de la recta
datos.plot.scatter(x='metro',y='precio')
plt.plot(x, y, '-r')
plt.ylim(0,datos['precio'].max()*1.1)
# plt.grid()
plt.show()
```

```python
# usando sklear para saber los valores optimos
from sklearn.linear_model import LinearRegression

# definiendo input y output
X_train = np.array(datos['metro']).reshape((-1, 1))
Y_train = np.array(datos['precio'])

# creando modelo
model = LinearRegression(fit_intercept=False)
model.fit(X_train, Y_train)

# imprimiendo parametros
print(f"intercepto (b): {model.intercept_}")
print(f"pendiente (m): {model.coef_}")
```
