```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("C:/Users/paula/OneDrive/Escritorio/UNAD/SEMESTRE_04_1604/ANALISIS_DE_DATOS/ETAPA_2/cars.csv",sep=",")
```

```python
data.plot.scatter(x="year",y="priceUSD")
plt.show()
```

```python
#ejmplo regresion linear
regresion = linear_model.LinearRegression()

#se agrega los datos en un array
years = data["year"].values.reshape((-1,1))

#se crea el modelo
modelo = regresion.fit(years, data['priceUSD'])

#imprime la interseccion con pruebas de request
print("Interseccion (b)", modelo.intercept_)
#imprime la pendiente
print("Pendiente (m)", modelo.coef_)
```

```python
entrada = [[1970],[1980],[1990],[2000],[2010]]
predicciones = modelo.predict(entrada)
print(predicciones)
```

```python
data.plot.scatter(x="year",y="priceUSD",label="datos originales")
plt.scatter(entrada,predicciones,color="green")
plt.plot(entrada,predicciones,color="grey", label="Linea de regresin")
plt.xlabel("year")
plt.ylabel("priceUSD")
plt.legend()
plt.show()
```
